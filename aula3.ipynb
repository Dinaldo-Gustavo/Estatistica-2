```python
usuarios = pd.read_csv(r'Usuarios.csv', sep=",", decimal=".", encoding="windows-1252")
print("Nome das colunas:")
print(usuarios.columns)
print("-" * 30)
print("Estrutura do DataFrame:")
usuarios.info()
print("-" * 30)
print("Resumo estatístico:")
print(usuarios.describe())
```

```python
freq = usuarios['grau_instrucao'].value_counts()
print(freq)
```

```python
freq_rel = usuarios['grau_instrucao'].value_counts(normalize=True)
print(freq_rel)
```

```python
p_freq_rel = usuarios['grau_instrucao'].value_counts(normalize=True) * 100
print(p_freq_rel)
```

```python
freq_estado = usuarios['estado_civil'].value_counts()
print(freq_estado)
```

```python
freq_rel_estado = usuarios['estado_civil'].value_counts(normalize=True)
print(freq_rel_estado)
```

```python
p_freq_rel_estado = usuarios['estado_civil'].value_counts(normalize=True) * 100
print(p_freq_rel_estado)
```

```python
freq['Total'] = freq.sum()
freq_rel['Total'] = freq_rel.sum()
p_freq_rel['Total'] = p_freq_rel.sum()
```

```python
tabela_final = pd.DataFrame({
    'freq': freq,
    'freq_rel': freq_rel,
    'p_freq_rel': p_freq_rel
})
print(tabela_final)
```

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv(r'mercadoimobiliario.csv', sep=';')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10,6))
ax = sns.boxplot(x='Bairro', y='Preco_m2', data=df, palette="Set2")

#sns.swarmplot(x='Bairro', y='Preco_m2', data=df, color='.25', size=4)

plt.title('Distribuição de Preço/m² por Bairro em São Paulo', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Preço por m² (R$)', fontsize=12)

plt.show()
```

```python
import pandas as pd

# Criar as Series (equivalente aos arrays do numpy)
anderson = pd.Series([100, 100, 100, 50, 50, 50])
patricia = pd.Series([75, 74, 76, 77, 75, 74])

# Calcular e imprimir as médias
media_anderson = anderson.mean()
media_patricia = patricia.mean()
print(f"Média de Anderson: {media_anderson}")
print(f"Média de Patrícia: {media_patricia}")
print("-" * 30)

# Calcular e imprimir as variâncias
var_anderson = anderson.var()
var_patricia = patricia.var()
print(f"Variância de Anderson: {var_anderson:.4f}")
print(f"Variância de Patrícia: {var_patricia:.4f}")
print("-" * 30)

# Calcular e imprimir os desvios padrão
dp_anderson = anderson.std()
dp_patricia = patricia.std()
print(f"Desvio Padrão de Anderson: {dp_anderson:.4f}")
print(f"Desvio Padrão de Patrícia: {dp_patricia:.4f}")
```
